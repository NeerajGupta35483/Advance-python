other classes by using instances of those classes as a instance

**Q5. What is the purpose of pseudoprivate attributes?**

**Ans:** Pseudoprivate attributes are also useful in larger frameworks
or tools, both to avoid introducing new method names that might
accidentally hide definitions elsewhere in the class tree and to reduce
the chance of internal methods being replaced by names defined lower in
the tree. If a method is intended for use only within a class that may
be mixed into other classes, the double underscore prefix ensures that
the method won't interfere with other names in the tree, especially in
multiple-inheritance scenarios

Pseudoprivate names also prevent subclasses from accidentally redefining
the internal method's names,
